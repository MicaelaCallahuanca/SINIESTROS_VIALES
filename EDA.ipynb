### Graficamos las variables numéricas

### Se realiza un histograma de las variables numéricas

```python
columnas_numericas = df_hechos.select_dtypes(include=['int', 'float']).columns

df_hechos[columnas_numericas].hist(bins=20, figsize=(15, 10), edgecolor='black')
plt.suptitle('Histograma de las columnas numéricas', y=1.02)
plt.show()
```

En el histograma de la columna 'N_VICTIMAS' podemos observar que se da mayormente una victima por hecho. En el histograma de la columna 'AAAA' podemos ver que en los años 2016 y 2018 se dío la mayor cantidad de siniestros viales. En la columna 'MM' vemos que los meses donde se produce los siniestros es el noviembre y diciembre (las fiestas). En 'DD', día, podemos ver que se ve que los accidentes ocurren mayormente los fines de semana. En la columna 'comuna' vemos que la comuna 1 es la que tiene mayor taza de siniestros viales.

### Se hace una matriz de correlación 

```python
dataNumerica = df_hechos.select_dtypes(include=['number']) 
corr = dataNumerica.corr().round(2)

plt.figure(figsize=(6,5))
sns.heatmap(corr, cmap="viridis", annot=True)
plt.title('Mapa de Correlación - Variables numéricas')
plt.show()
```

En esta matriz veremos los coeficientes de correlación de la variables numérica que indican la fuerza y dirección de la relación lineal entre las variables. Los valores pueden tomar valores entre -1 y 1.
-1 indica una correlación negativa perfecta, 
0 indica ausencia de correlación, 
1 indica que hubo una correlación positiva perfecta

## Gráficos

La moto es el transporte que más usó una victima a la hora del siniestro. El ir caminado es la segunda causa del accidente, seguida del auto

Los siniestros viales se dan en mayoría con una sola victima, se dan principalmente a las 7 de la mañana 

```python
ax= df_hechos['ACUSADO'].value_counts() \
        .plot(kind='bar', title='Métodos de transporte usados por los acusados', color='red')
ax.set_xlabel('transporte')
ax.set_ylabel('cantidad')
plt.show()
```

En este gráfico de barras representa los métodos de transporte usado por los acusados. Podemos observar que se compone mayormente por auto, seguida por los pasajeros de un transporte como, por ejemplo, micros de larga distancia. Seguida de cargas y objetos fijos

Los accidentes ocurren principalmente en avenidas, seguida por calle, autopista y Gral Paz

# Gráficos

### Histograma de las edades de las victimas

```python
ax = df_victimas['EDAD'].plot(kind='hist', 
                         title='Edades de las victimas', 
                         color='blue')
ax.set_xlabel('Edad')
plt.show()
```

Se observa que la victimas que estuvieron en los hechos son mayormente los que van de 20 a 40 años

### Gráfico de barras del rol de las victimas

```python
df_victimas['ROL'].value_counts().plot(kind='bar', color='blue')
plt.title('Frecuencia de roles en víctimas')
plt.xlabel('Rol')
plt.ylabel('Frecuencia')
plt.show()
```

El rol de las victimas con más accidentes son de conductor y peaton

### Gráfico de barras de tipo de victimas por años

```python
ca = ['VICTIMA', 'AAAA']
ax= df_hechos[ca].value_counts() \
        .plot(kind='bar', title='Tipo de victimas por años', color='green')
ax.set_xlabel('tipo de victima y año')
ax.set_ylabel('')
plt.show()
```

En el 2018 se registró más victimas de tipo peatón, seguida del 2016 con moto

### Histograma de edades de las victimas por sexos

```python
masculino = df_victimas[df_victimas['SEXO'] == 'MASCULINO']['EDAD'].dropna()

femenino = df_victimas[df_victimas['SEXO'] == 'FEMENINO']['EDAD'].dropna()

plt.figure(figsize=(12, 6))
plt.hist(masculino, bins=20, alpha=0.5, color='skyblue', label='Masculino', density=True)
plt.hist(femenino, bins=20, alpha=0.5, color='red', label='Femenino', density=True)
plt.title('Distribución de Edad de Víctimas por Sexo')
plt.xlabel('Edad')
plt.ylabel('')
plt.legend()
plt.show()
```

Se separaron los sexos en FEMENINO y MASCULINO, luego se realizó un histograma. En este gráfico podemos observar que el SEXO MASCULINO de entre 20 a 40 años son los que tuvieron más siniestros viales a comparación del SEXO FEMENINO

### Gráfico de barras de roles por sexo

```python
tabla_contingencia = pd.crosstab(df_victimas['ROL'], df_victimas['SEXO'])

ax = tabla_contingencia.plot(kind='bar', stacked=True, figsize=(12, 6), color=["lightcoral",'skyblue'])
plt.title('Distribución de Roles por Sexo')
plt.xlabel('Rol')
plt.ylabel('')


handles, labels = ax.get_legend_handles_labels()
plt.legend(handles=handles, labels=['FEMENINO', 'MASCULINO'], title='Sexo')

plt.xticks(rotation=0)
plt.show()
```

Podemos ver que una distición de roles de victimas por distinción de sexos. En el caso del rol de conductor se puede observar que el sexo masculino está más presente que el femenino, en el segundo lugar, el rol de peaton, se puede distinguir que también el sexo masculino es predominante que en el caso del sexo femenino
